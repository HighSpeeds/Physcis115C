```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wavefunction = lambda x,n: np.sqrt(2)*np.sin(n*np.pi*x)
x = np.linspace(0,1,1000)

for n in [2,4,6]:
    plt.plot(x,wavefunction(x,n)**2,label='n = {}'.format(n))

for n in [1,3,5,11]:
    plt.vlines(n/12,0,2,label='n = {}'.format(n),color="k")
plt.legend()
```

```python
plt.plot(x,wavefunction(x,2)**2,label='n = {}'.format(n))
plt.ylabel('$\\frac{E_{2}^1}{\\alpha}$')
plt.xlabel('location of the delta function divided by the length of the box')
plt.savefig('fig1.png')
```

```python
Lamdba = 1
omega = np.linspace(-10,10,1000)
Omega =2

plt.plot(omega,np.sqrt(omega**2+(Lamdba*Omega)**2),label='$\lambda = 1$',color='b')
plt.plot(omega,-np.sqrt(omega**2+(Lamdba*Omega/2)**2),color='b')

plt.plot(omega,omega,label='$\lambda = 0$',color='k')
plt.plot(omega,-omega,color='k')
plt.ylabel('Energy')
plt.xlabel('$\omega$')
plt.title('Energy vs $\omega$\n $\Omega=2$')
plt.legend()
plt.savefig('fig2.png')
```
