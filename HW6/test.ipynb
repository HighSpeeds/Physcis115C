```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
omega=np.linspace(-3,3,1000)

omega21 = 0
gamma = 1
Omega = (omega21-omega)**2 + gamma**2
plt.plot(omega, gamma**2/Omega**2)
plt.xlabel(r'$\omega$')
plt.ylabel(r'$|c_{2}(t)|^2$')
plt.title(r'$|c_{2}(t)|^2$ with $\omega_{21}=0$ and $\gamma=\hbar$')
plt.savefig('fig1.png')
```
