```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const
```

```python
L=1
hbar = const.hbar
m = const.m_e
n = np.arange(-3,4)
k = 2*np.pi*n/L
E = (hbar**2*k**2)/(2*m)
plt.plot(n,E,'o')
```

```python
beta = 2*np.pi*hbar/(2*m*L)
E_correction = -beta*hbar/L*k
```

```python
# Creating figure
fig = plt.figure()

# Plotting dataset_2
ax = fig.add_subplot(111)
# ax.plot(x, dataset_2, '-', label='dataset_2')
# ax.plot(x, dataset_3, '-', label='dataset_3')

# Creating Twin axes for dataset_1
ax2 = ax.twinx()

plt.sca(ax)
for i in range(len(n)):
    plt.plot([0,1],[E[i],E[i]+E_correction[i]],c = 'k')
plt.ylabel("initial energy values")
plt.yticks([])
plt.xticks([])
plt.title("Energy correction")

ax2.set_ylabel("final energy values")
plt.sca(ax2)
plt.yticks([])
plt.xticks([])
plt.savefig("2f.png")
```
