```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["font.family"] = "Times New Roman"
x_l1 = 1
x_l2 = 5

x_r1 = 7
x_r2 = 9

plt.plot([0,x_l1,x_l1,x_l2,x_l2,x_r1,x_r1,x_r2,x_r2,10],[0,0,-2,-2,0,0,-1,-1,0,0],color="black",linewidth=2)
plt.xticks([x_l2,x_r1],["$x_l$","$x_r$"])
plt.yticks([0],["$V_0$"])

def finite_well_wavefunction(x,L,E,V):
    if type(x)==np.ndarray:
        y = np.zeros(len(x))
        for i in range(len(x)):
            y[i] = finite_well_wavefunction(x[i],L,E,V)
        return y
    if np.abs(x)>L/2:
        return np.exp(-np.sqrt(2*E)*np.abs(x))
    if np.abs(x)<L/2:
        return np.cos(np.sqrt(2*(V-E))*x)/np.cos(np.sqrt(2*(V-E))*L/2)*np.exp(-np.sqrt(2*E)*L/2)

x = np.linspace(0,10,1000)
psi_l=finite_well_wavefunction(x-3,4,0.5,2)
plt.plot(x,psi_l/np.ptp(psi_l)*.5-1.5,color="black",linewidth=2,linestyle="-.",label = "$\psi_l$")
psi_r=finite_well_wavefunction(x-8,2,0.5,1)
plt.plot(x,psi_r/np.ptp(psi_r)*.5-0.75,color="black",linewidth=2,linestyle="--",label = "$\psi_r$")
plt.legend()
plt.xlabel("$x$")
plt.ylabel("$V(x)$")
plt.title("Potential Barrier")
plt.savefig("bardeen_tunneling.png",dpi=300)
```
